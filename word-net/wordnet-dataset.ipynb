```python
import random
from nltk.corpus import wordnet

# WordNet에서 모든 unique lemmas(단어 형태) 가져오기
lemmas = set(wordnet.all_lemma_names())

# 랜덤으로 775개의 단어 추출
random_words = random.sample(lemmas, 2338)

print(random_words)
```

```python
import random

def create_typo(word):
    if len(word) < 2:
        return word

    typo_type = random.choice(["insert", "delete", "replace"])
    position = random.randint(0, len(word) - 1)

    if typo_type == "insert":
        return word[:position] + random.choice("abcdefghijklmnopqrstuvwxyz") + word[position:]
    elif typo_type == "delete":
        return word[:position] + word[position + 1:]
    elif typo_type == "replace":
        return word[:position] + random.choice("abcdefghijklmnopqrstuvwxyz") + word[position + 1:]

# 사용 예시
word = random_words[2]
typo_word = create_typo(word)
print("Original:", word)
print("With Typo:", typo_word)
```

```python
train_typos = []
train_combination =[]
id_num = []
for each in random_words:
    train_typos.append(create_typo(each))
    train_combination.append(create_typo(each)+"<input>"+each )
    id_num.append(random_words.index(each))    
```

```python
import pandas as pd
df = pd.DataFrame()

df['original_words'] = random_words
df['typos'] = train_typos
df['name_line'] = train_combination
df['label'] = id_num
```

```python

df2 = pd.DataFrame(df)

# 랜덤하게 538개의 행 선택
random_rows = df2.sample(n=538)
```

```python
test_original_words = random_rows.original_words.tolist()
test_label = random_rows.label.tolist()
```
